## 6.ì‹œê³„ì—´ ë°ì´í„° ë‹¤ë£¨ê¸°

> ì‹œê³„ì—´ ë°ì´í„° : í–‰ê³¼ í–‰ì— ì‹œê°„ìˆœì„œê°€ ìˆëŠ” ë°ì´í„°(ì „ì œ : ì‹œê°„ìˆœìœ¼ë¡œ ì •ë ¬ ë˜ì–´ ìˆëŠ” ìƒíƒœ)

```python
from pykrx import stock
import datetime
import numpy as np
import pandas as pd

# ì˜¤ëŠ˜ ë‚ ì§œ ê°€ì ¸ì˜¤ê¸°
today = datetime.datetime.now()

# 1ë…„ ì „ ë‚ ì§œ ê³„ì‚°
one_year_ago = today - datetime.timedelta(days=365)

# ì£¼ê°€ ë°ì´í„° ë‹¤ìš´ë¡œë“œ ë° ê°„ë‹¨í•œ ì²˜ë¦¬
stock = stock.get_market_ohlcv_by_date(one_year_ago.strftime('%Y%m%d'), today.strftime('%Y%m%d'), "030200")
stock.reset_index(inplace = True)
stock = stock[['ë‚ ì§œ', 'ì¢…ê°€', 'ê±°ë˜ëŸ‰']]
stock.columns = ['Date', 'Close', 'Volume']
stock.head()
```

```python
# ğŸ¤”1. ì „ë‚ , ì „ì „ë‚  ì£¼ê°€ ë³€ìˆ˜ë¥¼ ì¶”ê°€í•˜ì‹œì˜¤. (.shift)
stock['shift1'] = stock['Close'].shift(1)
stock['shift2'] = stock['Close'].shift(2)

stock.head()
```

```python
# ğŸ¤”2. ì „ì „ë‚  ëŒ€ë¹„ ì „ë‚  ì£¼ê°€ ì¦ê° ë³€ìˆ˜ ì¶”ê°€ (.diff)
stock['diff2'] = stock['shift1'].diff(1)
stock.head()
```

```python
# ğŸ¤”3. 2ë²ˆ ë³€ìˆ˜ê°€ + ì´ë©´ 1, ì•„ë‹ˆë©´ 0 (np.where)
stock['updown'] = np.where(stock['diff2'] > 0, 1, 0)
stock.head()
```

```python
# ğŸ¤”4. 3ì¼ ì´ë™í‰ê·  ì£¼ê°€ ë³€ìˆ˜ë¥¼ ì¶”ê°€í•˜ì‹œì˜¤.(rolling().mean())
stock['3rolling'] = stock['Close'].rolling(3).mean()
stock.head()
```
