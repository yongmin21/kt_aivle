## [실습2] 제대로 불러오기

- head(), tail() 메서드로 조회 결과를 볼 때, 세 번째 행부터 읽어와야 합니다.
- header=2를 지정하여 다시 불러와 seoul_people 데이터프레임으로 선언하세요.

```python
# 데이터 불러오기(세 번째 행부터)
seoul_people = pd.read_csv(path + '1.3 seoul_people_202407.csv',
                           header=2, 
                           sep='\t',
                           encoding='CP949')

# 확인
seoul_people
```

- seoul_people 데이터프레임이 '동별(2)', '남자', '여자', '소계.1' 열만 갖게 하세요.

```python
# '동별(2)','남자','여자','소계.1' 행만 갖게
# df = df[['열1', '열2', '열3']] 형태의 구문 사용

seoul_people = seoul_people[['동별(2)', '남자', '여자', '소계.1']]

# 확인
seoul_people
```

- 소계인 첫 번째 행은 의미가 없으니 제거하고, 인덱스를 초기화하세요.

```python
# 첫 번째 행 제거
# 참고: df.drop(인덱스) 형태 구문 사용 
seoul_people = seoul_people.drop(0, axis=0)

# 확인
seoul_people
```

```python
# 인덱스 초기화
# 참고: reset_index() 메서드 사용
seoul_people = seoul_people.reset_index(drop=True)

# 확인
seoul_people
```

- 다음과 같이 seoul_people 데이터프레임의 열 이름을 변경하세요.
    - '동별(2)' --> '자치구', '소계.1' --> '합계'

```python
# 열 이름 변경: '동별(2)' --> '자치구', '소계.1' --> '합계'
# 참고: rename() 메서드 사용
seoul_people = seoul_people.rename({'동별(2)' : '자치구',
                                    '소계.1' : '합계'},
                                   axis=1)

# 확인
seoul_people
```

## [실습4] 시각화 분석

- 다양한 열를 기준으로 그래프를 그려보고 인사이트를 도출해보세요.

```python
import seaborn as sns
```

```python
# 주석
tmp = seoul_people.sort_values(['합계'], ascending=False)
plt.figure(figsize=(15, 8))
sns.barplot(data=tmp,
            x='자치구', y='합계')

plt.axhline(y=tmp['합계'].mean(), c='k')
plt.grid()
plt.show()
```

```python
# 인구가 가장 많은 구는 송파구, 강서구, 강남구 순이다.
# 인구가 가장 적은 구는 중구, 종로구, 용산구 순이다.
```

```python
# 주석
tmp = seoul_people.copy()
tmp['남여비율'] = tmp['남자'] / tmp['합계']
tmp['남여비율편차'] = tmp['남여비율'] - tmp['남여비율'].mean()
tmp = tmp.sort_values(['남여비율편차'], ascending=False)

plt.figure(figsize=(15, 8))
sns.barplot(data=tmp,
            x='자치구', y='남여비율편차')

plt.grid()
plt.show()
```

```python
# 값이 클수록 남자 비율이 많다.
# 값이 작을수록 여자 비율이 많다.
# 남자/합계 - 평균(남자/합계)이니까.
```

```python
# 위 차트를 통해 알게된 사실을 정리해 보세요.

# 인구가 가장 많은 구는 송파구, 강서구, 강남구 순이다.
# 인구가 가장 적은 구는 중구, 종로구, 용산구 순이다.

# 금천구, 관악구, 구로구 등은 남자 비율이 더 많고
# 마포구, 서대문고, 은평구 등은 여자 비율이 더 많다.
# 근소하게 차이 나는 정도이다.
```

```python
# 인구가 가장 많다는건 주거지가 많다는거 아닐까 반대로 인구가 적은건 주거지가 적은 대신 상업건물이 많거나?
# 남자 비율과 여자 비율의 차이는 치안과 관련 있을 수도 있겠다.
```
