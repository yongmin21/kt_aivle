## [실습4] 시각화 분석

- 다양한 열를 기준으로 그래프를 그려보고 인사이트를 도출해보세요.

```python
import seaborn as sns
```

```python
column = ['한식 일반 음식점업', '커피전문점', '기타주점업']

rank_df = (seoul_business[['자치구', column[0]]]
           .rank()
           .astype(int))

rank_df
```

```python
# 주석
column = ['한식 일반 음식점업', '커피전문점', '기타주점업']
plt.figure(figsize=(15, 30))
for i in range(3):
    tmp = (seoul_business[['자치구', column[i]]]
           .sort_values([column[i]], 
                        ascending=False))
    
    ax = plt.subplot(3, 1, i + 1)
    sns.barplot(data=tmp,
                x='자치구', y=column[i],
                ax=ax)
    
    plt.grid()
    
plt.show()
```

```python
# 위 차트를 통해 알게된 사실을 정리해 보세요.

# 업종별로 유형이 다를 것이다. (한식 top 3들은 상업지구, 커피 top 3들은 상업지구 및 여가지구, 주점 top3들은 여가지구)
# 주점업이 야간 버스 수요가 많지 않을까

# ANOVA
```

```python
import scipy.stats as spst

# 한식, 커피, 기타주점, 일반 교과, 한식 육류요리
p1 = seoul_business['한식 일반 음식점업']
p2 = seoul_business['커피전문점']
p3 = seoul_business['기타주점업']
spst.f_oneway(p1, p2, p3)
```
