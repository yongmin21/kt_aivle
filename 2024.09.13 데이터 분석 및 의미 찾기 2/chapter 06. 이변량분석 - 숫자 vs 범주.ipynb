## **2.숫자 --> 범주**

### **(1) 시각화**

* 히스토그램을 Survived로 나눠서 그려봅시다.

```python
sns.histplot(x='Age', data = titanic, hue = 'Survived')
plt.show()
```

* kdeplot을 그려봅시다.  
* 두가지 방법이 있습니다.
    * ① kdeplot(   , hue = 'Survived') 
        * 생존여부의 비율이 유지된 채로 그려짐
        * 두 그래프의 아래 면적의 합이 1
    * ② kdeplot(   , hue = 'Survived', common_norm = False)
        * 생존여부 각각 아래 면적의 합이 1인 그래프
    * ③ kdeplot(   , hue = 'Survived', multiple = 'fill')
        * 나이에 따라 생존여부 **비율**을 비교해볼 수 있음. (양의 비교가 아닌 비율!)


① kdeplot( , hue = 'Survived)

* common_norm = True (기본값)

```python
sns.kdeplot(x='Age', data = titanic, hue ='Survived', fill=True)
plt.show()
```

② kdeplot( , hue = 'Survived, common_norm = False)

```python
sns.kdeplot(x='Age', data = titanic, hue ='Survived', 
            common_norm = False, fill = True)
# 두 개의 kde plot의 면적 합이 1이 아닌, 각각의 면적 합이 1이 되게
plt.show()
```

③ multiple = 'fill' 옵션

```python
sns.kdeplot(x='Age', data = titanic, hue ='Survived'
            , multiple = 'fill')
plt.axhline(titanic['Survived'].mean(), color = 'r')
plt.show()
```

```python
sns.histplot(x='Age', data = titanic, bins = 16
             , hue ='Survived', multiple = 'fill')
plt.axhline(titanic['Survived'].mean(), color = 'r')
plt.show()
```

다음의 관계에 대해 시각화 해 봅시다.

* [문1] Fare(운임) --> Survived

```python
target = 'Survived'
feature = 'Fare'
```

```python
sns.kdeplot(x=feature, data = titanic, hue =target, 
            common_norm = False, fill = True)
# 두 개의 kde plot의 면적 합이 1이 아닌, 각각의 면적 합이 1이 되게
plt.show()
```

```python
# 두 그래프가 만나는 지점 - 아래 그래프에서 전체 평균선과 만나는 지점
```

```python
sns.kdeplot(x=feature, data = titanic, hue =target
            , multiple = 'fill')
plt.axhline(titanic['Survived'].mean(), color = 'r')
plt.show()
```

```python
# 운임이 약 300 이상인 경우는 대부분의 비율이 생존한 것을 볼 수 있고 (그러나 kde 상에서는 수가 적다.)
# 운임이 0 ~ 약 180 까지는 요금이 올라갈수록 생존률도 높아지지만, 180 ~ 300 까지는 감소하는 것을 알 수 있다.
# 운임이 0 부근인 사람들은 대부분 죽었음을 알 수 있다.
```

```python
sns.histplot(x=feature, data = titanic, bins = 20
             , hue =target, multiple = 'fill')
plt.axhline(titanic[target].mean(), color = 'r')
plt.show()
```

```python
# kde 상에서는 추정치로 그려졌지만 histplot에서는 데이터가 없는 부분이 존재함을 알 수 있다.
```
